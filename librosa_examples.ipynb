```python
import librosa as lb
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load the audio file
y, sr = lb.load("DEAM_audio/2.mp3")
```

```python
# Plot the waveform
plt.figure(figsize=(14, 5))
lb.display.waveshow(y, sr=sr)
plt.title('Forma d\'onda')
plt.show()
```

```python
# Spectrogram calculation
D = lb.stft(y)

# Mel-Spectogram conversion
S = lb.feature.melspectrogram(y=y, sr=sr)

# MFCC calculation
mfcc = lb.feature.mfcc(y=y, sr=sr)

# Plot the spectrogram
plt.figure(figsize=(10, 4))
lb.display.specshow(lb.power_to_db(S, ref=np.max), y_axis='mel', x_axis='time')  # Mel is a scale for frequencies, ref is the reference power
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.tight_layout()

```
